### Let's get some simple data visualizations

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Age'], bins=20, kde=True, color= 'green')
plt.title('Age Distribution of Trainers')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Economic Status', hue = 'Team Rocket', data = df, palette = 'magma')
plt.title('Economic Status vs Team Rocket Affliation')
plt.xlabel('Economic Status')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Profession', y='Average Pokemon Level', data=df, palette='magma')
plt.xticks(rotation=45)
plt.title('Average Pokemon Level by Profession')
plt.show()
```

```python
def count_Team_Rocket_by_column_name(df, column_name):
    grouped = df.groupby(column_name)['Team Rocket'].value_counts().unstack(fill_value=0)
    grouped = grouped.reset_index()
    grouped = grouped[[column_name, 'Yes', 'No']]
    return grouped
```

```python
def plot_count_of_Team_Rocket_by_column_name(df, column_name):
    counts_df = count_Team_Rocket_by_column_name(df, column_name)

    plt.figure(figsize=(8,5))
    counts_df.plot(x = column_name, kind='bar', stacked=True, ax = plt.gca())
    plt.title('Team Affiliation by ' + column_name)
    plt.xticks(rotation=45)
    plt.xlabel(column_name)
    plt.ylabel('Count')
    plt.show()

```

```python
columns = df.columns[1::]
for col in columns:
    print(f'----- {col} -----')
    plot_count_of_Team_Rocket_by_column_name(df, col)
    print('\n' + '-'*100 + '\n')
```
