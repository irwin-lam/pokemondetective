### Let's get some simple data visualizations

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Age'], bins=20, kde=True, color= 'green')
plt.title('Age Distribution of Trainers')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Economic Status', hue = 'Team Rocket', data = df, palette = 'magma')
plt.title('Economic Status vs Team Rocket Affliation')
plt.xlabel('Economic Status')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Profession', y='Average Pokemon Level', data=df, palette='magma')
plt.xticks(rotation=45)
plt.title('Average Pokemon Level by Profession')
plt.show()
```

```python
plt.figure(figsize=(8,5))
df['Most Used Pokemon Type'].value_counts().plot.pie(autopct='%1.1f%%', color=sns.color_palette('magma'))
plt.title('Distribution of Most used Pokemon Types')
plt.ylabel('')
plt.show()
```
