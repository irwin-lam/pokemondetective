```python
import pandas as pd
import numpy as np
```

```python
path = 'data\pokemon_team_rocket_dataset.csv'
df = pd.read_csv(path)
```

```python
df.head()
```

```python
df.drop(columns=['ID'], inplace=True)
```

```python
df.head()
```

```python
def changetoBool(columns, df):
    for col in columns:
        df[col] = df[col].map({True: 1, False:0})
    return df
```

```python
columns = ['Is Pokemon Champion', 'Rare Item Holder', 'Charity Participation']
update_df = changetoBool(columns, df)
```

```python
update_df.head()
```

```python
update_df.tail()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
```

```python
categorical_cols = update_df.select_dtypes(include=['object']).columns
```

```python
label_encoders = {}
for col in categorical_cols:
    le = LabelEncoder()
    update_df[col] = le.fit_transform(update_df[col])
    label_encoders[col] = le
```

```python
update_df.tail()
```

```python
#remove the last 1000 because they are unknown. We need to train test split the 4000 values
update_df = update_df.head(4000)
```

```python
target = 'Team Rocket'
X = update_df.drop(columns=[target])
Y = update_df[target]
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size =.2, random_state=42)
```

```python
models = {
    'Logistic Regression': LogisticRegression(),
    'Random Forest': RandomForestClassifier(),
    'Gradient Boosting': GradientBoostingClassifier(),
    'SVM' : SVC()
}
```

```python
for name, model in models.items():
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    print(f'----- {name} -----')
    print(f'Accuracy: {accuracy_score(y_test, y_pred):.4f}')
    print(classification_report(y_test, y_pred))
    print('\n' + '-'*35 + '\n')
```
